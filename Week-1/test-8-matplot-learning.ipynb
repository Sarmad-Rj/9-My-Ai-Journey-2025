```python
data = [1, 2, 2, 3, 3, 3, 4, 4, 5, 6, 6, 7]

# Frequency
from collections import Counter
freq = Counter(data)

x = list(freq.keys())
y = list(freq.values())

plt.plot(x, y, marker='o')
plt.title("Frequency Polygon")
plt.xlabel("Data")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Sample data

x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 50]
z = [5, 15, 25, 35, 45]

ax.plot3D(x, y, z, color='red')
ax.set_title("3D Line Plot")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

# make data
x = np.linspace(0, 10, 100)
y = 4 + 1 * np.sin(2 * x)
x2 = np.linspace(0, 10, 25)
y2 = 4 + 1 * np.sin(2 * x2)

# plot
fig, ax = plt.subplots()

ax.plot(x2, y2 + 2.5, 'x', markeredgewidth=2)
ax.plot(x, y, linewidth=2.0)
ax.plot(x2, y2 - 2.5, 'o-', linewidth=2)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4, 5])
y = np.array([10, 12, 8, 14, 7])

plt.plot(x, y, color='black', label='Line')

# Condition: y > 10
plt.fill_between(x, y, where=(y > 10), color='green', alpha=0.4, label='y > 10')

# Optional: fill where y <= 10
plt.fill_between(x, y, where=(y <= 10), color='red', alpha=0.4, label='y ≤ 10')

plt.title("Conditional Fill")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.legend()
plt.grid(True)
plt.show()
```
